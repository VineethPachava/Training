```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Tone.csv")
df
```

```python
np_array = df["SOUNDLEVEL"]
fourier =  np.fft.fft(np_array)
plt.plot(df["SOUNDLEVEL"])
```

```python
len(fourier)
```

```python
arr =fourier.conjugate()*fourier
arr = np.sqrt(arr)
arr = np.real(arr)
print(np.histogram(arr,bins = [1,2,3,4]))
```

```python
%matplotlib qt5
plt.hist(arr,df["Time  "])
```
