```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("TONE_1.csv")
```

```python
np_array = df["SOUNDLEVEL"]
fourier =  np.fft.fft(np_array)
plt.plot(df["SOUNDLEVEL"])
```

```python
len(fourier)
```

```python
arr =fourier.conjugate()*fourier
arr = np.sqrt(arr)
arr = np.real(arr)
```

```python
%matplotlib qt5
plt.plot(arr)
print(arr)
```

```python
bins = []
np.digitize(arr,bins)


```
