```python
df1 = df.loc[:,["LINEAR_ACCELERATION_X","LINEAR_ACCELERATION_Y","LINEAR_ACCELERATION_Z","SOUND_LEVEL"]]
```

```python
plt.plot(df["Time"] , df1.LINEAR_ACCELERATION_Z.rolling(2000).mean(),'g')
plt.plot(df["Time"] , df1.LINEAR_ACCELERATION_X.rolling(2000).mean(),'b')
plt.plot(df["Time"] , df1.LINEAR_ACCELERATION_Y.rolling(2000).mean(),'r')
plt.plot(df["Time"] , df1.SOUND_LEVEL.rolling(2000).mean(),'r')
plt.show()
```
