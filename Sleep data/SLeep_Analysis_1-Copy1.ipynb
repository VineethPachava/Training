```python
# mean = pd.concat([df.ix[:,i:i+3].mean(axis=1) for i in range(6,9,3)], axis=1)
ACC_X = np.array([])
ACC_Y = np.array([])
ACC_Z = np.array([])
for i in range(len(df['LINEAR_X'])):
    ACC_X = np.append(ACC_X, float(
        df['LINEAR_X'][i].tolist()))
    ACC_Y = np.append(ACC_Y, float(
        df['LINEAR_Y'][i].tolist()))
    ACC_Z = np.append(ACC_Z, float(
        df['LINEAR_Z'][i].tolist()))
mean = np.sqrt(np.square(ACC_X)+np.square(ACC_Y)+np.square(ACC_Z))
mean
```

```python
m = pd.DataFrame(mean)
# mean_normalized = (m - m.mean())/(m.max()-m.min())
m_bins  = int(len(m)/480)
m_arr = np.array_split(mean,m_bins)
sleep_matrix = []
m_arr
```

```python
sleep_matrix = []
for i in range(m_bins):
    m_arr[i] = (m_arr[i] - m_arr[i].mean())/(m_arr[i].max()-m_arr[i].min())
    bin_mean = m_arr[i].mean()
    bin_std = m_arr[i].std()
    threshold = 0.
    count_awake = 0
#     print(bin_mean,bin_std)
    for val in m_arr[i]:
        if val >= threshold:
            count_awake += 1
#             print(val,threshold)

    if (count_awake)/len(m_arr) > 0.40:
        sleep_matrix.append(1)
    elif (count_awake)/481 <= 0.40:
        sleep_matrix.append(0)
#         print(threshold)
print(sleep_matrix)
```
